# Concatenating

## Import Library

```python
import pandas as pd
```

## Create Pandas DataFrame

```python
df1=pd.DataFrame({'Dates' :['2020-01-01', '2020-01-02', '2020-01-03', '2020-01-04',
               '2020-01-05', '2020-01-06'] ,\
                'ID': [1,2,3,4,5,6]})

df2=pd.DataFrame({'Names':['A','B','C','D','E','F'],\
                  'ID':[4,5,6,7,9,8],\
                'Designation':['Manager','Employee','CEO','CTO','Manager','Employee']}) 
```

```python
df1
```

```python
df2
```

## Concat Pandas DataFrames

```python
concat = pd.concat([df1,df2],sort=False)
concat
```

## Concat DataFrames along Axis 

```python
concat_axis = pd.concat([df1,df2],sort=False,axis=1)
concat_axis
```

## Concat Pandas DataFrames with New Index

```python
append_ignore_index=df1.append(df2,sort=False ,ignore_index=True)
append_ignore_index
```
